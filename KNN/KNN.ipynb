## Import Packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Read Data

```python
df = pd.read_csv('./data/11_1_51_data_Temperature_Humidity.csv', delimiter=',')
df.columns=['T','H','Class']
```

## Plot

```python
labels, index = np.unique(df["Class"], return_inverse=True)
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)
sc=ax1.scatter(df['T'], df['H'], marker='o', c=index, alpha=1.0, s=100, edgecolor='black')
ax1.legend(sc.legend_elements()[0], labels)
plt.show()
```

# 3D Data

## Import Packages

```python

from mpl_toolkits import mplot3d

#  matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

## Plot

```python

#
df = pd.read_csv('./data/11_1_52_data_3D.csv', delimiter=',')

df.columns=['x','y','z']
x = df['x']
y = df['y']
z = df['z']

#
fig = plt.figure(figsize=(10, 10), dpi=100)
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, c='red', marker='o', edgecolors='k', s=100);
#ax.scatter3D(x2, y2, z2, c='green', marker='o', edgecolors='k');
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(0, 45)
plt.draw()
#plt.pause(36000)
```
